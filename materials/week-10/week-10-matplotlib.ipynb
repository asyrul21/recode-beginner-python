```python
# import library
# https://matplotlib.org/stable/tutorials/introductory/sample_plots.html
# https://matplotlib.org/stable/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py
import matplotlib.pyplot as plt
```

```python
# simple plot
# revenues over 12 months
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"]
revenues = [1000, 1200, 1500, 1100, 1300, 1600, 1550, 1650, 1700, 1650, 1750, 1800]

# plot
plt.plot(months, revenues)

# add title
plt.title("Revenue for 2020")

# # labels
plt.ylabel('Revenue (Thousands)')
plt.xlabel('2020')
```

```python
# multiplots
# import numpy
import numpy as np

# create evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# multi bar plot using the Figure instance
# https://www.tutorialspoint.com/matplotlib/matplotlib_bar_plot.htm
import numpy as np
import matplotlib.pyplot as plt

data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]

X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure

ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)
```
